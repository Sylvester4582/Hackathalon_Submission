```python
X=train_features.drop('respondent_id', axis=1)
y=train_labels.drop('respondent_id', axis=1)
```

```python
test=pd.read_csv('test_set_features.csv')
```

```python
test['income_poverty']=test['income_poverty'].replace({'<= $75,000, Above Poverty': 2, '> $75,000': 1, 'Below Poverty': 0})

test['marital_status']=test['marital_status'].replace({'Married': 1, 'Not Married': 0})

test['rent_or_own']=test['rent_or_own'].replace({'Own': 1, 'Rent': 0})

test['employment_status']=test['employment_status'].replace({'Employed': 2, 'Not in Labor Force': 0, 'Unemployed': 1})

test['age_group']=test['age_group'].replace({'65+ Years': 70, '55 - 64 Years': 60, '45 - 54 Years': 50, '35 - 44 Years': 40, '18 - 34 Years': 26})

test['education']=test['education'].replace({'College Graduate': 3, 'Some College': 2, '12 Years': 1, '< 12 Years': 0})

test['race']=test['race'].replace({'White': 3, 'Black': 2, 'Hispanic': 1, 'Other or Multiple': 0})

test['sex']=test['sex'].replace({'Male': 1, 'Female': 0})
```

```python
X_test=test.drop('respondent_id', axis=1)
```

```python
num_features = X.select_dtypes(include=['number']).columns.tolist()
cat_features = X.select_dtypes(include=['object']).columns.tolist()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
```

```python
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)
    ]
)
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', OneVsRestClassifier(LogisticRegression(max_iter=1000)))
])
```

```python
model.fit(X,y)
```

```python
y_pred=model.predict_proba(X_test)
```

```python
y_pred
```

```python
proba_df = pd.DataFrame(
    data=y_pred,
    columns=['h1n1_vaccine', 'seasonal_vaccine'],
    index=test['respondent_id']
)
```
